```python
import sqlite3
import pandas as pd
```

```python

# Read the CSV file into a DataFrame
table1 = pd.read_csv('Data1')
table1
```

```python

# Connect to a SQLite database
conn = sqlite3.connect('mydatabase.db')

# Save the DataFrame to a SQLite table
table1.to_sql('table1', conn, if_exists='replace', index=False)
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("SELECT * FROM table1")
data = cursor.fetchall()

for row in data:
    print(row)
```

```python
cursor.execute("PRAGMA table_info(table1)")
data = cursor.fetchall()

for row in data:
    print(row)
```

1. Write a query to give  details of hotels city is Sydney and  ratings is more than 5.

```python
cursor.execute("SELECT * FROM table1 WHERE [City Name] = 'Sydney'AND [Number of reviews] > 5")
data = cursor.fetchall()

for row in data:
    print(row)
```

2. Write a query  to find the location with the most reviews.


```python
cursor.execute('SELECT [Hotel Name], COUNT(*) AS [Number of reviews] FROM table1 GROUP BY [Hotel Name] ORDER BY [Number of reviews] DESC LIMIT 1')
data = cursor.fetchall()

for row in data:
    print(row)
```

3. Write a  query that  will retrieve all rows and sort them in descending order of user ratings.


```python
cursor.execute("SELECT * FROM table1 ORDER BY [Users ratings] DESC")
data = cursor.fetchall()

for row in data:
    print(row)
```

4. Write a query that  retrieves rows where the star rating is greater than or equal to 4.0 and sorts them in descending order of review score.


```python
cursor.execute("SELECT * FROM table1 WHERE [Star rating] >= 4.0 ORDER  BY [Review score] DESC")
data = cursor.fetchall()

for row in data:
    print(row)
```

5. Write a query to count the number of hotels in each city.


```python
cursor.execute("SELECT [City Name], count([Hotel Name]) FROM table1 GROUP BY [City Name]")
data = cursor.fetchall()

for row in data:
    print(row)
```

```python
# Read the CSV file into a DataFrame
table3 = pd.read_csv('Data3')
table3
```

```python

# Connect to a SQLite database
conn = sqlite3.connect('mydatabase.db')

# Save the DataFrame to a SQLite table
table3.to_sql('table3', conn, if_exists='replace', index=False)
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("PRAGMA table_info(table3)")
data = cursor.fetchall()

for row in data:
    print(row)
```

1. Write SQL query to retrieve hotels that offer double bed rooms and sort them by price in descending order.


```python
cursor.execute('SELECT * FROM table3 WHERE [Double bed] = 1 ORDER BY [Prices] DESC')
data = cursor.fetchall()

for row in data:
    print(row)
```

2. Write a SQL query to calculate the average price for each room type.


```python
cursor.execute('SELECT [Room Type], avg([Prices]) FROM table3 GROUP BY [Room Type]')
data = cursor.fetchall()

for row in data:
    print(row)
```

3. Write a query to find details of hotels with user ratings above 5 as a  threshold and that offer free WiFi. (Use table 1 and table 2) 


```python
cursor.execute('SELECT * FROM table1 INNER JOIN table2 on [Hotel Name] = [Hotel Name] WHERE [Review score] > 5.0 and [Wifi] = 1')
data = cursor.fetchall()

for row in data:
    print(row)
```

4. Write a query to find hotels that offer both family rooms and non-smoking rooms(Use table 1 and Table 2)


```python
cursor.execute('SELECT * FROM table1 inner join table2 on ([Hotel Name]) = ([Hotel Name]) where [Family rooms] = 1 and [Non Smoking Rooms]= 1')
data = cursor.fetchall()

for row in data:
    print(row)
```

5. Write a query to calculate the average price for each room type (Use table2 and Table 3)


```python
cursor.execute('SELECT [Hotel Name], [Room Type], avg([Prices]) from table3 GROUP BY [Hotel Name], [Room Type]')
data = cursor.fetchall()

for row in data:
    print(row)
```
